```python
df = df.sort_values(by=['AccountID', 'TransactionDate']).reset_index(drop=True)
```

```python
df["PreviousTransactionDate"] = df.groupby("AccountID")["TransactionDate"].shift(1)
df["PreviousTransactionDate"] = df["PreviousTransactionDate"].fillna(df["TransactionDate"])
```

```python
df['TransactionDate']=pd.to_datetime(df['TransactionDate'],errors='coerce')
df['PreviousTransactionDate']=pd.to_datetime(df['PreviousTransactionDate'],errors='coerce')
```

```python
def age_group(age):
    if age <=30:
        return "Young"
    elif age<=50:
        return "Middle"
    else:
        return "Senior" 
        
```

```python
df['AgeGroup']=df['CustomerAge'].apply(age_group)
```

```python
df.head()
```

```python
#  High Amount (> 99th percentile)
amount_threshold = df['TransactionAmount'].quantile(0.99)  # ~$1333 in sample
df['HighAmount'] = df['TransactionAmount'] > amount_threshold

# High Login Attempts (> 3)
df['HighLogins'] = df['LoginAttempts'] > 3


df['RapidTransaction'] = False
for account, group in df.groupby('AccountID'):
    time_diffs = group['TransactionDate'].diff().dt.total_seconds()
    df.loc[group.index[1:], 'RapidTransaction'] = time_diffs[1:] < 60


duration_threshold = df['TransactionDuration'].quantile(0.95)
df['UnusualDuration'] = df['TransactionDuration'] > duration_threshold



df['is_suspicious'] = (
    df['HighAmount'] |
    df['HighLogins'] |
    df['RapidTransaction'] |
    df['UnusualDuration']
).astype(int)
```

```python
df = df.drop(columns=['HighAmount', 'HighLogins', 'RapidTransaction', 'UnusualDuration'])
```
